# xticks
- tells matplotlib where to align/stick (specified in the sequence)

````python
x = np.random.randint(0, 100, 100)
plt.title("Age group and their count")
plt.xlabel("Age")
plt.ylabel("count")
y = [1,12,20, 59, 100]
plt.hist(x, y, edgecolor="white") # sequence serves as limiters
plt.xticks(y) # makes the axis limited to the sequence
plt.show()
````

````python
x = np.random.randint(0, 100, 100)
plt.title("Age group and their count")
plt.xlabel("Age")
plt.ylabel("count")
plt.hist(x, bins="auto", edgecolor="white") 
plt.show()
````

````python
# range
x = np.random.randint(0, 100, 100)
plt.title("Age group and their count")
plt.xlabel("Age")
plt.ylabel("count")
plt.hist(x, bins="auto", range=(20,50), edgecolor="white")  # range defines the lower and upper limit. It excludes the outliers in a data set
plt.show()
````

# formula for density is
- raw count/(total count * bind width)

````python
# weights
plt.title("Age group and their count")
plt.xlabel("Age")
plt.ylabel("count")
x = [1,23,25,34]
y= [1,11,21,35]
plt.hist(x, y, edgecolor="white", weights=[2,2,2,2], ) 
plt.grid(axis="y")
plt.show()
````

````python
# weights
plt.title("Age group and their count (cumulative)")
plt.xlabel("Age")
plt.ylabel("count")
x = [1,2,3,4,5]
y= [1,2,3,4,5, 6]
plt.hist(x, y, edgecolor="white", cumulative=True, ) # cumulatively plot the histogram
plt.grid(axis="y")
plt.xticks(y)
plt.show()
````

# define functions for getting the binwidth

````python
# rawcount / (len * (range/binCount))
def binwidth( arr, rowcount, bins=10):
  baba = ((max(arr)-min(arr))/bins)
  return rowcount/(len(arr)*baba)
````

# define function for graphing density 

````python
def densityHist(x, flag=False):
  plt.hist(x, density=flag, edgecolor="white")
  plt.grid(axis="y")
  plt.xlabel("AGE")
  # plt.ylabel("AGE")
  plt.show()
````

````python
# another formula for bin width is max -min/bin
def anotherFormula(arr, binNum):
  return (max(arr) - min(arr))/ binNum
````

````python
months = ["Jan", "Feb", "Mar", "April", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [4500, 5200, 4800, 5800, 6000, 6200, 5700, 5300, 5000, 6100, 6700, 7000]
customers = [120, 140, 130, 150, 160, 170, 155, 145, 138, 165, 180, 190]
````

````python
plt.plot(months, sales)
plt.title("Trend of sales over the months")
plt.xlabel("months")
plt.ylabel("sales")
plt.show()
````

````python
plt.bar(months, customers)
plt.title("Number of customers each month")
plt.xlabel("months")
plt.ylabel("customers")
plt.show()
````

````python
plt.scatter(sales, customers)
plt.xlabel("Sales")
plt.ylabel("Customers")
plt.show()
# conclusion: there is positive correlation between the sales and number of customers
````
